```python
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont

img = Image.open('form/.jpg')
texts_and_positions = {
        "{}".format("123456789"): (370, 1070),  # House Air Bill # 
        "{}".format("2022-10-10"): (370, 1110),   # Expected Date of delivery 
        "Him Chan Lee": (370, 1180), # Shippers Signature 
        "SC": (370, 1230)            # Shippers Name & Title 
    }

font = ImageFont.load_default()

#     .
draw = ImageDraw.Draw(img)
font_color = (0, 0, 0)  #   (RGB)
for text, position in texts_and_positions.items():
    draw.text(position, text, font=font, fill=font_color)

#  
plt.imshow(img)
plt.axis('on')  #  .
plt.show()
 
```
