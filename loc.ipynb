```python
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont

img = Image.open('form/김승태.jpg')
texts_and_positions = {
        "{}".format("123456789"): (370, 1070),  # House Air Bill #의 좌표
        "{}".format("2022-10-10"): (370, 1110),   # Expected Date of delivery의 좌표
        "Him Chan Lee": (370, 1180), # Shippers Signature의 좌표
        "SC": (370, 1230)            # Shippers Name & Title의 좌표
    }

font = ImageFont.load_default()

# 텍스트와 좌표에 텍스트를 이미지에 추가합니다.
draw = ImageDraw.Draw(img)
font_color = (0, 0, 0)  # 텍스트 색상 (RGB)
for text, position in texts_and_positions.items():
    draw.text(position, text, font=font, fill=font_color)

# 이미지 표시
plt.imshow(img)
plt.axis('on')  # 축을 숨깁니다.
plt.show()
 
```
